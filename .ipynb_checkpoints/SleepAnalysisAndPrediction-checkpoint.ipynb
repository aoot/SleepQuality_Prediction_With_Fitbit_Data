## Methods
---

### Data Source:
[![DOI](https://zenodo.org/badge/DOI/10.5281/zenodo.53894.svg)](https://doi.org/10.5281/zenodo.53894)

During this analysis, I gathered a Fitbit timeseries data that was made available through the Amazon Web Services (AWS) crowdsourcing serice, Mechnical Turk. The dataset is then provided by the researchers as open dataset.

### Data Description:
The dataset consist of 33 volunteers' Fitbit data over a period of 31 days (April 12, 2016 - May 12, 2016). Most of the data are recorded at a per-second resolution, with a combination of measured data (i.e. heart rate, steps, etc.), aggregated or calculated metrics (i.e. activity intensity, MET, etc.), user reported data (i.e. calory intake, weights).

### Preparing the Data:
To ready the data for model analysis, missing data were imputed with either zero or the mean of the subset, and resampled to a resolution of per-day. Most metrics (i.e. steps, activity intensity, minutes asleep, etc.) were aggregated by summing the entire day's record or aggregated as a mean (i.e. heart rate). Some featuring engineering were done to extract further insights specifically on the date-time of each observation entry. The result is a dataset with 744 observations and 8 features to splice upon.

### Exploratory Analysis: 
The dataset is then analyzed for correlation, statistical distribution, trend analysis over time (31 days), and aggregated to some of the features (i.e. grouped-by day-of-week for day-to-day trends. We find correlation between data and aggregated feature (i.e. steps and activityIntensity) as Fitbit's algorithm may have considered steps in its reporting of activityIntensity.

### Machine Learning Model Analysis and Building:
We used four different supervised machine learning models to fit the data and used $R^2$ as the primary metrics of performance measurement between the models. The four models used are Multi-Linear Regression, Decision Tree Regression, AdaBoost (Decision Tree Enesemble), and Support Vector Machine. Each of the model (exclude linear regression) has its hyperparameter tuned using scikit-learn's RandomizedSearchCV to find the optimal hyperparamter through stochastic fitting and cross-validation of the training set. 

## Discussion:
---
Of the four different models, the SVM ($R^2 = 0.172$) has the worst performance measured in $R^2$ value, and AdaBoost with hyperparameter tuning at the individual tree level and ensembled level provides the best performance ($R^2 = 0.636$). 

Through multi-linear regression model's coefficient aboslute value ranking, we see that weekday is the strongest predictor of a person's sleep quality (measured in sleepDepth).

Further, during the EDA stage, some interesting patterns were found, such as the confirmation that "hump day" indeed is a low point of the week as interpreted through shallow sleep, low levels of activity in terms of intensity and step counts.

## Conclusion:
---
Circling back to the question of how to get better sleep through actional steps, it seems that the most actional step that is within control of an individual is to add some high intensity workout during mid-weed to counterbalance the various unmeasured impacts that leads to shallower sleep during mid-week.

Further, an AdaBoost model can be potentially used to predict sleepDepth of an individual. This model can be further elaborate with higher resolution data to provide a reminder or warning system for a fitness wearable device throughout the day to remind the user to increase add a workout into the rest of their day or increase their acitivity level.

The model is only able to achieve an $R^2$ of around 60%, and I believe this is due to limited observation and participants in the dataset. With only 33x participants over a span of 31 days lessing the missing data, we are working with a limited dataset.

### Helper Function Definition

```python
### HELPER FUNCTION: Map participant Id to 0-32 (Since we know there are 33 volunteer/participants)

# ## List of participants
# participantList = minIntensities.Id.value_counts().index
# ## Create a dictionary map
# participantMap = {}  # Dict placeholder
# for idx, participantId in enumerate(participantList):
#     participantMap[participantId] = idx


## Function to remap ID in dataframe
def remapParticipants(data=None, idColName="id"):
    """Remap participant ID column to the specified mapping."""
    
    participantMap = {1927972279: 0, 2026352035: 1, 2873212765: 2, 4558609924: 3, 1624580081: 4, 4388161847: 5, 8053475328: 6, 2022484408: 7, 2320127002: 8, 8877689391: 9, 8378563200: 10, 7086361926: 11, 4445114986: 12, 4020332650: 13, 6962181067: 14, 1844505072: 15, 4702921684: 16, 5553957443: 17, 4319703577: 18, 8583815059: 19, 1503960366: 20, 5577150313: 21, 1644430081: 22, 3977333714: 23, 8792009665: 24, 6290855005: 25, 6117666160: 26, 6775888955: 27, 7007744171: 28, 3372868164: 29, 8253242879: 30, 2347167796: 31, 4057192912: 32}
    data[idColName] = data[idColName].replace(participantMap)
    data = data.sort_values(by=idColName)
    return data
```

```python
### HELPER FUNCTION: Function to add missing time records back
## TODO:
## - [ ] Improve imputing-by-mean to a day's range

## Libraries
import numpy as np
from datetime import datetime
import pandas as pd

def addMissingTimeAndResample(df=None, dateTimeCol=None, colToAgg=None, impute=0,
                              resampleFreq="D", agg="sum",
                              startDatetime="2016/04/12 00:00:00", 
                              endDatetime="2016/05/12 23:59:59"):
    """Add the missing seconds data and resample to hourly.
    
    The data are recorded at second level resolution, however some seconds are missing.
    Thus, this function iterates through each participant ID and fill in the missing seconds.
    The missing value is filled with np.nan.
    
    """
    
    ## Set DatetimeIndex for new index
    measurement_start = datetime.strptime(startDatetime, "%Y/%m/%d %H:%M:%S")
    measurement_end = datetime.strptime(endDatetime, "%Y/%m/%d %H:%M:%S")
    experiment_dateRange = pd.date_range(start=measurement_start, end=measurement_end, freq='S')
    experiment_dateRange = pd.DatetimeIndex(experiment_dateRange)
    
    partList = df["id"].unique()
    df_new = pd.DataFrame()  # Return object placeholder
    
    ## Loop through each participant to reindex the datetime to fill in missing time
    for part in partList:
        #print(f"Working on ID={part}")  # Debug message
        df_temp = df.loc[df["id"]==part]  # Subset to one participant
        if impute=="mean":
            fill_value = df_temp[colToAgg].mean()
        elif (type(impute) == int)|(type(impute) == float):
            fill_value = impute
        df_temp = df_temp.set_index(dateTimeCol).reindex(experiment_dateRange, fill_value=fill_value)  # Fill missing values with np.nan
        df_temp = (df_temp.resample(resampleFreq).sum() if agg=="sum" else df_temp.resample(resampleFreq).mean())  # Resample to hourly; use sum so not to have floating point errors
        df_temp['id'] = part  # Reset Id labels
        df_temp["hour"] = df_temp.index.hour
        df_temp["weekday"] = df_temp.index.weekday
        df_temp = df_temp.reset_index()  # Reset the index
        df_temp = df_temp.rename(columns={"index":"datetime"})  # Rename the resetted datetime column from index to og name
        df_new = pd.concat([df_new, df_temp])
        
    return df_new
```

### Basic Multi-Linear Regression

- Highest $R^2$ is 0.444
- By looking at the best estimator (found through cv with the highest $R^2$), we see that `weekday` varaiable has the largest coefficient, then `activityIntensity`.
- This is saying that in a basic multi-linear regression, `weekday` has the most predictive power (by a large margin to the second most predictive variable).
- Application of this may be to dynamically adjust one's activity to compensate those the weekdays with low sleep depth; in other words, if we know certain weekdays we have shallow sleep, we can attempt to improve it by improving activityIntensity.
- ??? Is weekday most predictive because of the interactions between weekday and activityIntensity? >>> Look into further by including interaction terms.

```python
### Basic linear regression to see the variable interactions

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
import numpy as np

linear_reg = LinearRegression()
linear_reg_cv_results = cross_validate(linear_reg, 
                                       X_train, y_train, 
                                       scoring=["r2","neg_root_mean_squared_error"],
                                       return_train_score=True, 
                                       return_estimator=True,
                                       cv=20)
```

```python
## Plot the cv train/test score
fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
## Plotting R^2
ax[0].set_title("Train/Test R^2 Score | Multi-Linear Regression")
ax[0].set_ylabel("R^2 Score")
ax[0].set_xlabel("CV iteration")
ax[0].plot(np.arange(1, 21, 1), linear_reg_cv_results["train_r2"], label="Train R^2")
ax[0].plot(np.arange(1, 21, 1), linear_reg_cv_results["test_r2"], label="Test R^2")
ax[0].text(1, 0.10, f"Train R2 (mean): {round(linear_reg_cv_results['train_r2'].mean(), 3)}")
ax[0].text(1, 0.07, f"Test R2 (mean): {round(linear_reg_cv_results['test_r2'].mean(), 3)}")
ax[0].legend()
## Plotting RMSE
ax[1].set_title("Train/Test RMSE Score | Multi-Linear Regression")
ax[1].set_ylabel("RMSE Score")
ax[1].set_xlabel("CV iteration")
ax[1].plot(np.arange(1, 21, 1), -linear_reg_cv_results["train_neg_root_mean_squared_error"], label="Train RMSE")
ax[1].plot(np.arange(1, 21, 1), -linear_reg_cv_results["test_neg_root_mean_squared_error"], label="Test RMSE")
ax[1].text(1, 260, f"Train RMSE (mean): {round(-linear_reg_cv_results['train_neg_root_mean_squared_error'].mean(), 3)}")
ax[1].text(1, 250, f"Test RMSE (mean): {round(-linear_reg_cv_results['test_neg_root_mean_squared_error'].mean(), 3)}")
ax[1].legend()
fig.show()
```

```python
## Analyze the Coefficients

## Finding the index with the max R^2
values = linear_reg_cv_results["test_r2"]
index_max = max(range(len(values)), key=values.__getitem__)

## Retrieve the estimator using this index
best_estimator = linear_reg_cv_results['estimator'][index_max]
best_test_r2 = linear_reg_cv_results['test_r2'][index_max]
best_estimator_coef = dict(zip(best_estimator.feature_names_in_, best_estimator.coef_))
best_estimator_coef = dict(sorted(best_estimator_coef.items(), key=lambda item: item[1], reverse=True))  # Sort by significance

print()
print(f"Best estimator coefficient: \n\t{best_estimator_coef}")
print()
print(f"Best R^2 Score: \n\t{best_test_r2}")
print()

## NOTES:
## - `weekday` has the largest coefficient, then acitivtyIntensity
```

### Basic Decision Tree + Parameter Tuning
- Start with basic decision tree and default hyperparamter; the perfomance is bad at $R^2 = 0.465$
- Thus we use RandomizedSearchCV to tune the hyperparameter of the decision tree and improved the $R^2 = 0.600$
    - Best parameter found using cv=10, n_iter=100: {'min_samples_split': 0.1, 'min_samples_leaf': 3, 'max_leaf_nodes': 10, 'max_depth': 10}

```python
### Basic decision tree, with default hyperparameters

## Libraries
from sklearn.tree import DecisionTreeRegressor

## Instantiating the tree
tree = DecisionTreeRegressor().fit(X_train, y_train)

## R^2 score
tree_r2 = tree.score(X_test, y_test)
print(f"R^2 Score of default decision tree hyperparameters: {tree_r2}")
```

```python
### Hyperparameter tune the decision tree

## Libraries
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import pandas as pd

## Hyperparameter Tuning with RandomSearchCV
param_dist = {"max_depth":np.arange(1, 20),
              "min_samples_split":np.arange(0.1, 1.0, 0.1),
              "min_samples_leaf":np.arange(1, 10),
              "max_leaf_nodes":np.arange(2, 20)}
tree_tune = RandomizedSearchCV(estimator=tree, 
                               param_distributions=param_dist,
                               n_iter=100,
                               cv=10,
                               verbose=1).fit(X_train, y_train)

## tree_tune results
tree_tune_cv_results = pd.DataFrame(tree_tune.cv_results_)

## Best parameters
tree_tune_best_param = tree_tune.best_params_
tree_tune_best_score = tree_tune.best_score_  # R^2, DecisionTreeRegressor's default score

print()
print(f"CV tuned tree best parameters: \n\t{tree_tune_best_param}")
print()
print(f"CV tuned tree with best R^2 Score: \n\t{tree_tune_best_score}")
print()
```

### AdaBoost (DT Ensemble) with Hyperparameter Tuning

We compared AdaBoost prior and after hyperparamter tuning, and each of them is fitted with either the basic DT or the best DT. 

We found that prior to hyperparameter tuning, the basic tree has an R^2 that is greater than the best tree.

However, after hyperparamter tuning, the best tree has a greater R^2 value then that of the basic tree; and I believe this is because we are able to tune the hyperparameter to the two different base estimator types.

#### AdaBoost prior to Hyparameter Tuning
There are two types of adaBoost ensembles here. The "basic_tree" is based on the simple decision tree with default parameters from the previous fitting; and the second type (e.g. "best_tree") is based on decision tree with tuned hyperparameters found in previous steps.

We see that the "basic_tree" version of the AdaBoost has an $R^2$ value of 0.624 and the "best_tree" version of the AdaBoost has n $R^2$ value of 0.534.

```python
### Train AdaBoost Ensemble with default hyperparameters (prior to hyperparameter tuning)

## Libraries
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

## Train AdaBoost with basic tree (default parameter)
basic_tree = tree
## Instantiate AdaBoost model
adaBoost_basic_tree = AdaBoostRegressor(base_estimator=basic_tree,
                                        n_estimators=30,
                                        learning_rate=1.0,
                                        loss="linear").fit(X_train, y_train)
print(f"Basic Tree AdaBoost R^2 Score: {adaBoost_basic_tree.score(X_test, y_test)}")
```

```python
## Train AdaBoost with best tree found preivously
best_tree = tree_tune.best_estimator_
## Instantiate DT ensemble
adaBoost_best_tree = AdaBoostRegressor(base_estimator=best_tree, 
                                       n_estimators=30, 
                                       learning_rate=1.0,
                                       loss="linear").fit(X_train, y_train)
print(f"Best Tree AdaBoost R^2 Score: {adaBoost_best_tree.score(X_test, y_test)}")


## Pickle the model
# pickleToFile(adaBoost_basic_tree, "AdaBoost_Basic_Tree")
# pickleToFile(adaBoost_best_tree, "AdaBoost_Best_Tree")
```

#### AdaBoost with Hyparameter Tuning
There are two types of adaBoost ensembles here. The "basic_tree" is based on the simple decision tree with default parameters from the previous fitting; and the second type (e.g. "best_tree") is based on decision tree with tuned hyperparameters found in previous steps.

We see that the "basic_tree" version of the AdaBoost has an $R^2$ value of 0.618 and the "best_tree" version of the AdaBoost has n $R^2$ value of 0.636.

```python
### RandomizedSearchCV on the basic tree AdaBoost vs best tree AdaBoost
## NOTES:
## - The best tree is found using RandomizedSearchCV in earlier steps to find the 
##   best parameters for the AdaBoost base tree estimator.

## Libraries
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
import numpy as np
import pandas as pd

## Parameters to tune
param_dist = {"n_estimators":np.arange(1, 200, 1),
              "learning_rate":np.arange(0.1, 30, 1), 
              "loss":["linear", "square", "exponential"]}

## Basic Tree Estimator for the RandomizedSearchCV
basic_tree = adaBoost_basic_tree

## Tune the AdaBoost on a basic DT
adaBoost_basic_tree_tuning = RandomizedSearchCV(basic_tree, 
                                                param_distributions=param_dist, 
                                                n_iter=10,
                                                cv=5).fit(X_train, y_train)

# display(pd.DataFrame(adaBoost_basic_tree_tuning.cv_results_))
print(f"Best Param from adaBoost using basic tree: \n\t{adaBoost_basic_tree_tuning.best_params_}")
print(f"Best R^2 from adaBoost using basic tree: {adaBoost_basic_tree_tuning.best_score_}")
print()
```

```python
## Best Tree Estimator for the RandomizedSearchCV
best_tree = adaBoost_best_tree

## Tune the AdaBoost on a best DT
adaBoost_best_tree_tuning = RandomizedSearchCV(best_tree, 
                                                param_distributions=param_dist, 
                                                n_iter=10,
                                                cv=5).fit(X_train, y_train)

# display(pd.DataFrame(adaBoost_best_tree_tuning.cv_results_))
print(f"Best Param from adaBoost using best tree: \n\t{adaBoost_best_tree_tuning.best_params_}")
print(f"Best Score from adaBoost using best tree: \n\t{adaBoost_best_tree_tuning.best_score_}")
print()
```

### Support Vector Machine (Regressor)

We compare the two scenarios of fitting a SVM to different base estimators. The first SVM has a base estimator that has yet to fit to the training data, and the second SVM has its base estimator fit to the data prior to hyperparameter tuning. 

The results yields the same best_parameters set and statistically negligible difference in R^2. 
This result makes sense as even if the base_estimator is fit prior to hyperparameter tuning, the training dataset are the same thus there are no information gain here.

However, compared to the case above with AdaBoost, because AdaBoost is an ensemble, the ensembled model would be different if the base estimator is different. Thus, we see a performance boost with higher R^2 when fitting ensembling the hyperparameter tuned base decision trees prior to tuning the parameter of the AdaBoost model.

#### SVR using Estimator that hasn't been fit

```python
## Libraries
from sklearn.svm import SVR
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
## Test if fitting estimator prior to RandomizedSearchCV makes a difference

## No prior fit - The SVR has yet to be fit prior to passing to RandomziedSearchCV
svr_reg = SVR()
param_dist = {"kernel":["liner", "poly", "rbf"],
              "degree":np.arange(2, 4, 1),   # Only when kernel is "poly", ignored by others
              "C":np.arange(1, 5, 1)}
svr_reg_tune = RandomizedSearchCV(svr_reg, 
                                  param_distributions=param_dist, 
                                  n_iter=10,
                                  cv=5,
                                  verbose=1).fit(X_train, y_train)

## Pickle the model
# pickleToFile(svr_reg_tune, "SVR_REG_TUNE")  # Pickle the file so not to have to retrain

print()
print(f"SVR best parameters: \n\t{svr_reg_tune.best_params_}")
print()
print(f"SVR best R^2 score: \n\t{svr_reg_tune.best_score_}")  # R^2 of the best parameter SVR (kernel=='rbf')
print()
```

#### SVR using Estimator that has been fit already

```python
## Prior fit - The SVR has been fit to the train dataset prior to RandomizedSearchCV
svr_reg = SVR().fit(X_train, y_train)
param_dist = {"kernel":["liner", "poly", "rbf"],
              "degree":np.arange(2, 4, 1),   # Only when kernel is "poly", ignored by others
              "C":np.arange(1, 5, 1)}

## Tune the parameter
svr_reg_tune_prior_fit = RandomizedSearchCV(svr_reg, 
                                  param_distributions=param_dist, 
                                  n_iter=10,
                                  cv=5,
                                  verbose=1).fit(X_train, y_train)

## Pickle the model
# pickleToFile(svr_reg_tune_prior_fit, "SVR_REG_TUNE_PRIOR_FIT")  # Pickle the SVR model

print()
print(f"SVR Best parameters: \n\t{svr_reg_tune_prior_fit.best_params_}")
print()
print(f"SVR Best R^2 Score: \n\t{svr_reg_tune_prior_fit.best_score_}")
```
